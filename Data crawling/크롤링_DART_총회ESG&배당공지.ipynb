# 1) dart ESG 공시검색결과 크롤링

```python
path = "chromedriver.exe"
url = 'https://dart.fss.or.kr/dsab007/main.do'

wd = webdriver.Chrome(path)
wd.maximize_window() #창 최대화

wait = WebDriverWait(wd, 5) #no such window 문제 방지하기 위해 wait 생성

# 사이트로 접속
wd.get(url)
wd.implicitly_wait(3)


# [공시통합검색] 옵션 변경 ---------------------------------------------------
## 1) 전체 → 본문내용으로 변경
select = Select(wd.find_element_by_id('option')) #select 사용할 것
select.select_by_value('contents') #본문내용 선택
time.sleep(1)

## 2) 검색어 입력: ESG
keyword = wd.find_element_by_xpath('//*[@id="keyword"]')
keyword.send_keys('ESG')
wd.implicitly_wait(1)

## 3) 기간 입력 (2021.01.01 ~ 2021.12.31)
start = wd.find_element_by_xpath('//*[@id="startDate"]')
start.clear()
start.send_keys('20210101')
end = wd.find_element_by_xpath('//*[@id="endDate"]')
end.clear()
end.send_keys('20211231')
time.sleep(1)

## 4) 공시유형: 기타공시에 check.
checkbox = wd.find_element_by_xpath("//input[@type='checkbox' and @id='dspType5']")
#driver.execute_script("arguments[0].scrollIntoView(true);", elem)
checkbox.click()

## 5) 검색 버튼 클릭
wd.find_element_by_xpath('//*[@id="keywordBtn"]').click()
time.sleep(3)



# 크롤링 준비
## 기업 이름 담을 list 준비
name_list = []
## 파일 명 담을 list 준비
file_list = []
## 날짜 담을 list 준비
date_list = []

# 페이지 별 긁어오기
for i in range(50):
    
    if not (i == 0):
        # 다음 페이지로 넘어가기.
        try:
            wd.find_element_by_xpath(f'//*[@href="javascript:search({i+1});"]').click()
            time.sleep(3)
        # 에러 뜨는 경우, 더이상 넘어갈 다음 페이지가 없음.
        except:
            break #for문 종료
    
    names = wd.find_elements_by_xpath('//*[@id="listContents"]/div[2]/table/tbody/tr/th/span/a')
    files = wd.find_elements_by_xpath('//*[@id="listContents"]/div[2]/table/tbody/tr/th/a')
    dates = wd.find_elements_by_xpath('//*[@id="listContents"]/div[2]/table/tbody/tr/td[3]')
    
    new_names = [name.get_attribute('innerText') for name in names]
    new_files = [file.get_attribute('innerText') for file in files] 
    new_dates = [date.get_attribute('innerText') for date in dates]

    name_list.extend(new_names)
    file_list.extend(new_files)
    date_list.extend(new_dates)
```

```python
wd.quit()

ESG_industry = pd.DataFrame({'회사명':name_list, 'file':file_list, 'date':date_list})
ESG_industry = ESG_industry[-ESG_industry['file'].str.contains("정정")] 
ESG_industry.to_excel('G crawling/result/DART총회_ESG_2021.xlsx', index=None, encoding='cp949')
```

# 2) dart 보고서 현금배당결정 관련된 변수 크롤링

```python
path = "chromedriver.exe"
url = 'https://dart.fss.or.kr/dsab007/main.do'

wd = webdriver.Chrome(path)
wd.maximize_window() #창 최대화

wait = WebDriverWait(wd, 5) #no such window 문제 방지하기 위해 wait 생성

# 사이트로 접속
wd.get(url)
wd.implicitly_wait(3)


# [공시통합검색] 옵션 변경 ---------------------------------------------------
## 1) 전체 → 본문내용으로 변경
select = Select(wd.find_element_by_id('option')) #select 사용할 것
select.select_by_value('report')
time.sleep(1)

## 2) 검색어 입력: 관련 문구 삽입 
keyword = wd.find_element_by_xpath('//*[@id="reportName"]')
keyword.send_keys('현금ㆍ현물배당결정//현금배당결정')
wd.implicitly_wait(1)

## 3) 기간 입력 (2021.01.01 ~ 2021.12.31)
start = wd.find_element_by_xpath('//*[@id="startDate"]')
start.clear()
start.send_keys('20210101')
end = wd.find_element_by_xpath('//*[@id="endDate"]')
end.clear()
end.send_keys('20211231')

## 4) 조회건수 100으로 설정
select = Select(wd.find_element_by_id('maxResultsCb'))
select.select_by_value('100')
time.sleep(2)

## 5) 검색 버튼 클릭 (엔터로 입력)
keyword.send_keys(Keys.ENTER)
time.sleep(3)



# 크롤링 준비
## 이름 담을 list 준비
CASH_name_list = []
CASH_file_list = []
CASH_person_list = []
CASH_date_list = []
# note_list = []

# 페이지 별 긁어오기
for i in range(100):
    
    if not (i == 0):
        # 다음 페이지로 넘어가기.
        try:
            wd.find_element_by_xpath(f'//*[@href="javascript:search({i+1});"]').click()
            time.sleep(3)
        # 에러 뜨는 경우, 더이상 넘어갈 다음 페이지가 없음.
        except:
            break #for문 종료
            
    
    list1 = wd.find_elements_by_xpath('//*[@id="tbody"]/tr/td[2]/span/a')
    list2 = wd.find_elements_by_xpath('//*[@id="tbody"]/tr/td[3]/a')
    list3 = wd.find_elements_by_xpath('//*[@id="tbody"]/tr/td[4]')
    list4 = wd.find_elements_by_xpath('//*[@id="tbody"]/tr/td[5]')
    list5 = wd.find_elements_by_xpath('//*[@id="tbody"]/tr/td[6]/span')
    
    new_names = [a.get_attribute('innerText') for a in list1]
    new_files = [a.get_attribute('innerText') for a in list2]
    new_person = [a.get_attribute('innerText') for a in list3]
    new_dates = [a.get_attribute('innerText') for a in list4]
    
    CASH_name_list.extend(new_names)
    CASH_file_list.extend(new_files)
    CASH_person_list.extend(new_person)
    CASH_date_list.extend(new_dates)
```

```python
wd.quit()

while 'IR' in CASH_name_list:
    CASH_name_list.remove('IR') #기업 이름에 붙어있는 'IR' 문자열 제거
    
CASH_industry = pd.DataFrame({'회사명':CASH_name_list, 'file':CASH_file_list, 'person':CASH_person_list, 'date':CASH_date_list})
CASH_industry = CASH_industry[-CASH_industry['file'].str.contains("정정")] 
CASH_industry.to_excel('G crawling/result/DART배당공지_2021.xlsx', index=None, encoding='cp949')
```
