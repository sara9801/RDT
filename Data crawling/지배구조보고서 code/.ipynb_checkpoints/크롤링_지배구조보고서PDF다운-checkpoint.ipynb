## PDF 다운로드 code

셀레니움은 네트워크 환경의 영향이 매우 크기 때문에 오류가 발생할 수 있다.  
오류는 WebDriverWait와 time.sleep를 사용하여 방지했으나, PDF를 반복적으로 다운받는 과정에서 과부하가 일어나 코드가 중단되는 현상이 자주 일어났다.  

따라서 코드를 실행할 때 예기치 못한 오류가 발생한다면 관련 코드에 대해 time.sleep의 시간을 증가시켰다.  
time.sleep은 비효율적이나 가장 단순하기 때문에 사용하기엔 편리하다.    

만약 코드를 실행하다 오류가 발생한다면, 아래의 코드에 포함된 time.sleep의 시간을 늘려줌으로써 해결할 수 있다.

```python
# 빈 DF 만들기, 이후에 row 추가할 것
GR_name = pd.DataFrame({'name':[], 'pdf':[]})
```

```python
path = "chromedriver.exe"
url = 'https://esg.krx.co.kr/contents/02/02040000/ESG02040000.jsp'



# 옵션 변경, 바로 pdf 다운로드할 수 있게끔.
options = webdriver.ChromeOptions()
options.add_experimental_option('prefs', {"download.prompt_for_download": False, #To auto download the file
                                          "plugins.always_open_pdf_externally": True}) #Do not show PDF directly in chrome

wd = webdriver.Chrome(path, options=options)
wd.maximize_window() #창 최대화

wait = WebDriverWait(wd, 10) #no such window 문제 방지하기 위해 wait 생성

# 사이트로 접속
wd.get(url)
wd.implicitly_wait(3)

# 원하는 년도 year=2020년도일 때.
# 조회기간 2021년도로 설정 ---------------------------------------------------------
if year==2020:
    ## 조회기간 변수 설정
    date = ['20210101', '20211231'] #start, end

# 원하는 년도 year=2021년도일 때.
# 조회기간 2022년도로 설정 ---------------------------------------------------------
if year==2021:
    ## 조회기간 변수 설정
    date = ['20220101', '20220927'] #start, end


# 조회기간 관련 변수 갖고오기
date_path = wd.find_elements_by_xpath('//*[@class="schdate"]')
for i in range(2):
    ## 조회기간 clear하여 공백으로 만들기
    date_path[i].clear()
    ## 조회기간 변수 입력
    date_path[i].send_keys(date[i])

## 조회버튼 click
lookup = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, 'sch_btn')))
lookup.click()
time.sleep(3)
# ----------------------------------------------------------------------------------


## 페이지 별 추출
a = 1
while True: # -----------------------------------------------------------------------------------------------------
    
    print(str(a)+" page is crawled...")
    a+=1 # 안전장치
    if a>40: break
    
    # 현재 탭 저장
    main_window  = wd.current_window_handle
    
    # 회사명 및 링크 추출
    time.sleep(5)
    com_list = wd.find_elements_by_xpath('//*[@name="com_nm"]')
    
    
    # 회사별 추출 ------------------------------------------------------------------------------
    for com in com_list:
        
        ## 각 회사별 [회사명, 지배구조보고서 pdf명] 담을 list.
        name = []
        
        ## 회사명은 list에 넣고, 회사 이름 클릭하여 링크로 이동
        name.append(com.get_attribute("innerText"))
        com.click()

        tabs = wd.window_handles #탭 2개로 확장
        wd.switch_to.window(tabs[1]) #2번째 탭으로 이동

        ## 첨부서류 -> 첫번째 요소 클릭
        select = Select(wd.find_element_by_id('attachedDoc')) #select 사용할 것
        sel_op = select.options
        sel_op.pop(0) #select option의 맨 처음인 '첨부문서선택' 삭제

        ## 만약 option의 text 값이 '신고서'를 포함하지 않는다면, 그 옵션의 value값을 integer로 변환하여 리스트에 반환.
        sel_value = [int(op.get_attribute("value")) for op in sel_op if '신고서' not in op.text]

        ## 최종 리스트의 길이가 1 = 기업지배구조 보고서만 남았음 
        if len(sel_value) == 1:
            select.select_by_value(str(sel_value[0]))

        ## 나왔던 기업지배구조 보고서 중 가장 최신 날짜를 가진 보고서의 value 추출
        else:
            final_value = str(np.sort(sel_value)[-1])
            select.select_by_value(final_value)

        ## 기업지배구조보고서.pdf 찾기
        wd.switch_to.frame('docViewFrm')
        time.sleep(1)
        
        try:
            ## 지배, 구조, 보고서 세 단어들을 모두 포함하는 파일을 가져온다.
            contain_str = "//a[contains(text(),'지배') and contains(text(), '구조') and contains(text(),'보고서')]"
            find_pdf = wd.find_elements_by_xpath(contain_str) 

            ## 만약 찾은 pdf 개수가 1이 맞으면 그 파일명 반환
            if len(find_pdf) == 1:
                find_pdf[0].click() #클릭해서 pdf 파일 다운로드
                name.append(find_pdf[0].text) #pdf명 name list에 저장

            ## 찾은 pdf 개수가 2개 이상이면 두 제목 중에 '정정'이라는 단어를 포함한 파일명을 반환
            elif len(find_pdf) > 1:
                titles = [i.text for i in find_pdf]
                find_text = [title for title in titles if '정정' in title]

                ### 만약 그 길이가 1개면, 클릭해서 파일 다운 & 이름 list에 추가
                if len(find_text) == 1:
                    find_pdf[ find_pdf.index(find_text[0]) ].click()
                    name.append(find_text[0])

                ### 길이가 0이면 정정이 문제가 아님. '신고서' 단어를 제외한 파일명을 반환
                elif len(find_text) == 0:
                    find_text = [title for title in titles if not '신고서' in title]
                    
                    #### 또 길이가 1인지 확인해서 파일 다운 & 이름 list에 추가
                    if len(find_text) == 1:
                        find_pdf[ find_pdf.index(find_text[0]) ].click()
                        name.append(find_text[0])
                    else:
                        name.append('ERROR')

                ### 그 이외의 경우는 리스트에 ERROR 반환
                else:
                    name.append('ERROR')
            
            else:
                ## 그 이외의 경우 = 찾은 pdf 개수가 애초에 0 = ERROR 반환
                name.append('ERROR')
        
        # 아예 지배, 구조, 보고서를 모두 포함한 pdf가 없는 경우. NON.
        except:
            name.append('NON')

        ## 창 닫고 기다리기
        wd.close()
        time.sleep(1)
        wait.until(EC.number_of_windows_to_be(1)) #윈도우 1개 될 때까지 waiting
        wd.switch_to.window(main_window) #원래 탭으로 돌아가기
        
        
        # 데이터 프레임 GR_name에 list=name을 마지막행으로 추가. 
        GR_name.loc[len(GR_name)] = name
        
    # ----------------------------------------------------------------------------------------------
    # 한 페이지 내의 회사별 pdf 다운 및 추출이 종료.
    
    
    # 다음 페이지로 이동.
    try:
        wd.find_element_by_xpath('//*[@class="next"]').click()
        time.sleep(3) #loading time
    
    #더 이상 넘어갈 페이지가 없을 시 종료
    except NoSuchElementException:
        time.sleep(10) #pdf 파일 모두 다운할 때까지 waiting
        print("crawling is done.")
        break


wd.quit()    
```

## 파일명 변경

GR_name에 저장된 pdf 파일명은 사실 실제 파일명과 다르다.  
예를 들어, [ 2020년도 기업지배구조 보고서.pdf ]라는 파일명을 가진 기업은 현대자동차와 롯데지주, 오뚜기이다.  


그러나 파일명이 같기 때문에, 실제 파일을 저장했을 때 롯데지주와 오뚜기는 DataFrame(약칭 DF)에 저장된 pdf 명과 다르게 저장된다.  
현대자동차 다음에 수집된 롯데지주는 [ 2020년도 기업지배구조 보고서 (1).pdf ]으로,  
그 다음에 수집된 오뚜기는 [ 2020년도 기업지배구조 보고서 (2).pdf ]으로 저장되는 것이다.


이를 맞춰주기 위해 다음과 같은 코드를 짜게 되었다.

### 1) pdf_adj이 같을 경우, 뒷번호에 numbering

```python
df = pd.read_csv('지배구조보고서/GR_name_adj.csv', encoding='cp949')
df.head()
```

```python
# pdf_adj에서 unique한 중복값을 추출
dup_name = df.pdf_adj.loc[df.pdf_adj.duplicated()].unique().tolist()
dup_name
```

```python
from collections import Counter
counter = Counter(df.pdf_adj)

for i, name in enumerate(dup_name):
    num = counter[name] - 1
    pdf_F = [name]
    
    for n in range(num):
        index = name.find('.pdf')
        pdf_F.append(name[:index] + ' (' + str(n+1) + ')' + name[index:])
    
    df.loc[df.pdf_adj == name, 'pdf_adj'] = pdf_F
    
# check
print('{}: pdf_adj의 중복행 개수'.format(df.pdf_adj.duplicated().sum()))
print('{}: 중복행이 제거된 데이터프레임 차원'.format(df.shape)) #아직 그대로.
```

### 2) name~pdf 모두 같을 경우, 중복행 제거

```python
dup = df.duplicated(subset=['name', 'pdf']).tolist()
df.loc[dup]

# 총 3개 기업이 'name' & 'pdf' 명이 같다.
```

```python
# 중복행 제거
df.drop_duplicates(subset=['name', 'pdf'], keep = 'first', inplace=True)

# check
print('{}: 중복행 개수'.format(df.duplicated(subset=['name', 'pdf']).sum()))
print('{}: 중복행이 제거된 데이터프레임 차원'.format(df.shape))
```

### 3) 증권회사의 연차보고서 구분하는 column 생성



```python
stock = []

for index, pdf in enumerate(df.pdf_adj):
    
    # 증권회사의 지배구조보고서: 지배구조 및 보수체계 연차보고서 
    # (우리에게 필요한 정보 없음)
    
    if ('보수체계' in pdf) or ('연차보고서' in pdf):
        stock.append(1)
    else:
        stock.append(0)
```

```python
df['broker'] = stock
df.tail()
```

```python
df['broker'].tolist().count(1)
```

### 4) 실제 파일명이 지정된 path에 존재하는지 확인.

```python
# 새로운 col에 수정된 이름 추가
# pdf_F: Final pdf name 이라는 의미.

## 파일 형식이 PDF가 아닌 파일은 직접 PDF로 변환하여 다시 저장.
## 이때 파일 형식명도 .docx에서 pdf로 변환해주었다.
## 파일 형식 이름이 대문자 PDF인 것도 편의를 위해 소문자 pdf로 변환하였다.
pdf_name = df.pdf_adj.tolist()
df['pdf_adj'] = [a.replace('[ ', "", 1).replace('.pdf ]', ".pdf").replace('.PDF ]', ".pdf").replace('.docx ]', '.pdf') for a in pdf_name]
df.rename(columns={'pdf_adj':'pdf_F'}, inplace = True)
```
