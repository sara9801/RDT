### pdf 추출 코드

```python
def extract_pdf_page(pdf_name, pdf_folder='pdf'):
    
    # 경고메세지 끄기
    warnings.filterwarnings(action='ignore')
    
    
    # Finding <key>
    Q1 = re.compile('주주총회 *4주 *전에 *소집공고 *실시') #1번 문항의 내용
    Q2 = re.compile('전자 *투표 *실시') #2번 문항의 내용

    Q15_1 = re.compile('중요 *정보에 *내부 *감사 *기구') #15번 문항의 일부 내용 1
    Q15_2 = re.compile('접근할 *수 *있는 *절차') #15번 문항의 일부 내용 2
    
    
    # Make a page_index DF. (return할 결과값.)
    page_index = pd.DataFrame({'start_page':[], 'end_page':[]})

    
    # for문 실행. 모든 pdf에 대해 페이지 추출.
    for index, name in enumerate(pdf_name):

        page = [-1, -1] #page 인덱스 삽입할 list 생성

        if (index+1) % 10 == 0:
            print(str(index+1) + "번째 pdf를 분석 중입니다...")

        #불러올 파일명
        file = '지배구조보고서/' + str(pdf_folder) + '/' + name

        # read the pdf
        pdf = PdfFileReader(open(file, 'rb'), strict=False) #읽기모드로 open
        num_pages = pdf.numPages #total pages

        # GOAL: '핵심지표 준수 현황' 관련된 문장을 포함한 페이지를 추출
        ## page 거꾸로 불러오는 이유: 핵심지표 준수 현황의 table은 대부분 pdf의 마지막에 위치.
        for i in range(num_pages, 0, -1):

            #만약 핵심지표 준수현황 작성 기준 페이지가 존재하면 이 페이지 넘어가고 index 초기화

            # page에서 text 추출
            text = pdf.pages[i-1].extractText()
            text = text.replace('\n', '') #줄바꿈 제거
            text = ' '.join(text.split()) #다중공백 제거

            if '핵심지표 준수현황 작성 기준' in text:
                page = [-1, -1] #페이지 초기화
                continue

            # 마지막 문항을 포함한 페이지
            if bool(re.search(Q15_1, text)) or bool(re.search(Q15_2, text)):
                page[1] = i

            # 첫번째 문항을 포함한 페이지
            if bool(re.search(Q1, text)) or bool(re.search(Q2, text)):
                page[0] = i    
                ## (**) 설명 하단에.
                if (page[1] == -1) or ((page[0]>page[1]) & page[1]>0):
                    page[1] = num_pages if (num_pages-i <= 4) else i+3
                page_index.loc[index] = page #DF에 page list 삽입
                break #페이지 모두 찾았으므로 stop.

            if i < 20:
                page_index.loc[index] = page #DF에 page list 삽입
                break # 핵심지표 준수 현황은 적어도 20 페이지 이상에서 나온다.
                      # 따라서 20페이지까지 도달하면 -1 삽입하고 다음 파일로 넘어감.

    
    
    # DF 변환 --------------------------
    ## 1) 데이터 타입 변환
    page_index = page_index.astype({'start_page':'int', 'end_page':'int'}) # data type floot->int로 전환
    
    ## 2) new column = industry_name에 pdf_name 삽입
    page_index['industry_name'] = pdf_name
    
    # ----------------------------------
    
    # 최종 DF 반환
    return(page_index)
    
    
    ## (**) if문은 두가지 안좋은 상황을 가정하고 있다.
    #        1) 앞문항(Q1, Q2) 찾았으나 마지막 문항은 못찾았을 경우: ex. page=[44, -1]
    #        2) page 인덱스가 방향이 반대일 경우: ex. page=[44, 43]
    #
    # 따라서 이러한 경우를 if문으로 찾고, 이런 일이 발생했을 땐 page[0]을 기준으로 page[1]를 재정의해준다.
    # 만약 total page 수가 page[0]와 많이 차이가 난다면 page[1]에는 page[0]+3 값을 넣어주고,
    # total page 수가 page[0]와 별 차이가 없다면 total page 수를 page[1]에 넣어주는 것이다.
    #
    # 이러한 아이디어는 핵심지표 준수 현황이 대부분 페이지 끝에 존재하므로,
    # 만약 첫번째 문항이 있는 페이지가 거의 끝페이지라면 / 마지막 문항은 페이지 끝이라고 가정하는 것이다.
    # 또한 첫번째 문항의 페이지는 44페이지이나 전체 페이지는 70페이지 가까이 된다면,
    # 이때는 핵심지표 준수 현황 표의 최대 페이지 수인 4페이지를 기준으로 삼고 page[0]+3을 마지막 문항이 존재하는 페이지로 간주한다.
```

## 준수여부 문항 찾는 함수 생성

#### Q라는 단어에 대한 문항을 찾는 함수

```python
def find_Q(Q):
    
    # 오류 방지를 위해 모든 특수문자 삭제
    new_keyind = df['핵심지표'].str.replace(pat=r'[^\w]',repl='', regex=True)
    key_series = df['준수여부_X'][new_keyind.str.contains(Q)]

    try:
        # 찾고자하는 핵심지표 없으면 list index로 꺼낼 때 error.
        key = key_series.tolist()[0]
        key = key.replace('\n', '').replace(' ', '') #줄바꿈, 띄어쓰기 삭제
        
        # 만약 데이터가 있으면(if len(key)>0) 체크되어 있는 것.
        ## 따라서 준수여부 X라는 뜻으로 0 삽입
        # 만약 데이터가 없으면(if len(key)==0), 준수여부 O에 체크되어 있다는 것.
        ## 따라서 준수여부 O라는 뜻으로 1 삽입 (준수=1, 준수안함=0)
        result = 1 if len(key) == 0 else 0
        return(result)
    
    # 이럴 땐 list에 -1 삽입
    except IndexError:
        return(-1)
    
    # 그밖의 에러에는 -9 삽입
    except:
        return(-9)
```

#### 준수현황(여부) 반환하는 함수
각 Q(준수여부 문항)에 대해 find_Q 함수를 실행하여 여부 반환한다.

```python
pattern_punctuation = re.compile(r'[^\w]')

def extract_comp(pdf_name, df):
    
    try:
        # data 넣을 새로운 리스트
        ## 우선 pdf_name 포함
        new_list = [pdf_name]

        # 만약 DF의 columns(header)에 핵심지표와 준수여부가 없다면, -99 삽입
        if not ('핵심지표' in df.columns.tolist())&('준수여부_O' in df.columns.tolist()):
            new_list.extend([-99 for i in range(15)])
            return(new_list)

    #         Q_list = ['소집', re.compile('전자투표'), '집중', '계획', #Q1~Q4: 주주 관련 핵심지표
    #                   '승계', re.compile('내부통제'), '분리', re.compile('집중투표제'), '방지', '초과', #Q5~Q10: 이사회
    #                   '교육', re.compile('내부감사부서'), '전문가', re.compile('분기별'), re.compile('중요정보')] #~Q15

        Q_list = ['소집', '전자투표', '집중일', '계획', #Q1~Q4: 주주 관련 핵심지표
                  '승계', '내부통제', '분리', '집중투표제', '방지', '초과', #Q5~Q10: 이사회
                  '교육', '내부감사부서', '전문가', '분기별', '정보'] #~Q15: 감사기구

        for q in Q_list:
            new_list.append(find_Q(q))
            
    except:
        new_list = [new_list[0]]
        new_list.extend([-999 for i in range(15)])
    
    return(new_list)
```

#### 데이터 프레임의 header(열 이름)이 되어줄 key 정의 함수

따라서 key가 잘 정의되도록 함수를 사용하여 보다 체계적으로 코드를 만들었다.  
아래의 Final code를 함께 보면 이해가 쉽다!

```python
def def_key(text):
    
    text_list = list(text)
    text_list = [pattern_punctuation.sub('', i) for i in text_list] #특수기호 제거
    text_list = [a.replace('구분', '') for a in text_list] #핵심지표와 구분 붙어있을 경우: 구분을 삭제
    comp_index = [i for i, text in enumerate(text_list) if "준수여부" in text]
    
    # 만약 길이가 1 초과 (2 이상)이라면 이 행이 준수여부를 포함한다는 뜻 → key로 설정할 것.
    if len(comp_index) > 1:
        
        ## 준수여부_O, 준수여부_X로 변경
        text_list[comp_index[0]] = '준수여부_O'
        text_list[comp_index[1]] = '준수여부_X'
        
        ## 최종 반환할 키 정의
        keys = tuple(text_list)
        
    else:
        keys = None
    
    
    return(keys)
```

```python
# return할 새로운 DF
new_df = pd.DataFrame({'pdf_name':[], 'Q1':[], 'Q2':[], 'Q3':[], 'Q4':[], 'Q5':[], 'Q6':[], 'Q7':[], 
                       'Q8':[], 'Q9':[], 'Q10':[], 'Q11':[], 'Q12':[], 'Q13':[], 'Q14':[], 'Q15':[]})

for index, pdf_name in enumerate(page_df['industry_name']):

    print(str(index+1) + "번째 docx에서 추출 중입니다.")
    
    # pdf_name을 docx_name으로 변경, 그리고 이 docx를 document로 불러온다.
    docx_name = '지배구조보고서/docx/' + pdf_name.replace('.pdf', '.docx')
    document = Document(docx_name)

    # doc에서 table들을 꺼내온다.
    # table은 1~4개가 될 수 있음 (2개 이상일 경우 table을 합친다.)
    tables = document.tables

    # 각 rows 삽입할 data 생성
    data = []
    
    # Dictionary 생성을 위한 keys
    keys = None
    

    # 모든 tables에 대해 수행: -------------------------------------------------------------
    for tab_i, table in enumerate(tables):
        for i, row in enumerate(table.rows):

            ## table의 각 cell에 대한 text를 불러온다.
            text = (cell.text for cell in row.cells)

            ## table의 첫번째 줄(header)을 key로 정의하고, 만약 두번째 테이블 이상이면 넘어감.
            if (i == 0)&(tab_i == None):
                ## key로 할 text가 준수여부 포함하고 있는지 확인하고,
                ## 포함하고 있으면 준수여부 포함한 행 중에, 앞에서부터 두개를 준수여부_O, 준수여부_X로 변경 (이유: 아래 하단**)
                ## 만약 포함하지 않으면 다음 row로 넘어간다. (key가 되기에 적합하지 않은 컬럼이라는 뜻.
                keys = def_key(text)
                continue
            
            if keys == None:
                keys = def_key(text)
                continue

            # dict 생성하고 data에 추가
            row_data = dict(zip(keys, text))
            data.append(row_data)
    # -------------------------------------------------------------------------------------

    # for문 종료. 모든 table에서 추출한 데이터 기반으로 데이터 프레임 생성
    df = pd.DataFrame(data)

    # 준수여부 list 반환하는 함수 실행 (extract_comp, find_Q)
    new_list = extract_comp(pdf_name, df)

    # 최종 결과를 DF에 삽입
    new_df.loc[index] = new_list


    # 위 과정에서 error 뜬 경우.
    # case 1. table 추출했으나 text가 아예 비어있는 경우 (ex. 2020년도 기업지배구조 보고서.docx)
    #except:
    #    new_list.extend([-999 for i in range(15)])
    #    new_df.loc[index] = new_list
    
    
    
    # ** 이유
    #준수여부가 여러 개면 가장 마지막 열에 대응됨. 그러나 2021년부터는 지배구조 보고서에 직전년도 준수여부까지 포함되면서 준수여부 행이 4개로 늘어나며 이럴 경우 직전년도의 준수여부 열만 남게 되어 데이터 자체에 오류가 생겨버린다. 따라서 앞에 우리가 필요로 하는 두 개의 컬럼(이번년도 준수여부 O, 이번년도 준수여부 X)에 준수여부_O, 준수여부_X라는 이름을 부여하여 이름을 특정시킨다.
```
